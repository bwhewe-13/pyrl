## Example of Q-Learning with Gymnasium Environment

```python
import gymnasium as gym
import matplotlib.pyplot as plt
import numpy as np

from pyrl import modeling 
```

```python
env = gym.make("FrozenLake-v1", is_slippery=False)

model = modeling.QLearning(env, gamma=0.9, alpha=0.1, epsilon=0.95)

train_rewards, train_steps = model.train(num_episodes=1000)
test_rewards, test_steps = model.test()

print(test_rewards)
```

```python
episodes = np.arange(1, len(train_rewards) + 1)

fig, ax = plt.subplots()
ax.scatter(episodes, train_steps, c="r", alpha=0.2, label="Training")
ax.legend(loc=0, framealpha=1)
ax.grid(which="both")
ax.set_xlabel("Episode Number")
ax.set_ylabel("Episode Length")
ax.set_title("Episode Length")
plt.show()
```
