Example of Q-Learning with Gymnasium Environments

```python
import gymnasium as gym
import matplotlib.pyplot as plt
import numpy as np

from pyrl.modeling import train
```

```python
env = gym.make("FrozenLake-v1", is_slippery=False)

model = train.QLearning(env)

train_rewards, train_lengths = model.train(episodes=1000, gamma=1.0, alpha=0.1)
test_rewards, test_lengths = model.test(num_episodes=1000)
```

```python
episodes = np.arange(1, len(train_rewards) + 1)

fig, ax = plt.subplots()
ax.plot(episodes, train_lengths, c="r", alpha=0.7, label="Training")
ax.plot(episodes, test_lengths, c="b", alpha=0.7, label="Testing")
ax.legend(loc=0, framealpha=1)
ax.grid(which="both")
ax.set_xlabel("Episode Number")
ax.set_ylabel("Episode Length")
ax.set_title("Episode Length")
plt.show()
```
